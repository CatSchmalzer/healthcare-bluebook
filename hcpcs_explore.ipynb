## Read in 1000 rows of Outpatient Charge data

```python
%%time

outpatient_df = pd.DataFrame()
chunk_nbr = 1

for chunk in pd.read_csv('../data/Provider_Outpatient_Hospital_Charge_Data_by_APC__CY2017.csv', 
                         chunksize = 1000):
    if chunk_nbr == 1:
        outpatient_df = outpatient_df.append(chunk)
        chunk_nbr += 1
    else: break
outpatient_df.head()
```
